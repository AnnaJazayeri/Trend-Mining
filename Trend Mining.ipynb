### Part 3: Consider the following directed graph. Write a program that uses Networkx to answer the followings:

    • Number of nodes and edges
    • Average in degree
    • Average out degree
    • Degree Centrality
    • Betweenness Centrality
    • Closeness Centrality

```python
# pip install networkx
import networkx as nx

# edges
edges = [('A', 'B'), ('B', 'C'), ('C', 'E'), ('E', 'D'), ('D', 'B'), ('E', 'F')]

# Creating the directed graph
G = nx.DiGraph()

# Add edges to the graph
G.add_edges_from(edges)

# Number of nodes and edges
num_nodes = G.number_of_nodes()
num_edges = G.number_of_edges()

print("\n\nNumber of nodes:", num_nodes)
print("Number of edges:", num_edges)

# Average in degree
avg_in_degree = sum(dict(G.in_degree()).values()) / num_nodes
print("\n\nAverage in degree:", avg_in_degree)

# Average out degree
avg_out_degree = sum(dict(G.out_degree()).values()) / num_nodes
print("Average out degree:", avg_out_degree)

# Degree Centrality
degree_centrality = nx.degree_centrality(G)
print("\n\nDegree Centrality:", degree_centrality)

# Betweenness Centrality
betweenness_centrality = nx.betweenness_centrality(G)
print("\n\nBetweenness Centrality:", betweenness_centrality)

# Closeness Centrality
closeness_centrality = nx.closeness_centrality(G)
print("\n\nCloseness Centrality:", closeness_centrality)
print("\n\n")
```

```python
import matplotlib.pyplot as plt

# edges
edges = [('A', 'B'), ('B', 'C'), ('C', 'E'), ('E', 'D'), ('D', 'B'), ('E', 'F')]

# Create a directed graph
G = nx.DiGraph()

# Add edges to the graph
G.add_edges_from(edges)

# Draw the graph
pos = nx.spring_layout(G) 
nx.draw(G, pos, with_labels=True, node_size=800, node_color="skyblue", font_size=12, font_weight="bold", arrowsize=20)
plt.title("Directed Graph")
plt.show()
```
